# **2. Verifikasi Kualitas Data**

**Verifikasi kualitas data** adalah proses memastikan data layak dianalisis dengan mengecek:
- Missing values (nilai kosong)
- Inconsistent values (nilai tidak konsisten)
- Duplicate values (data duplikat)
- Outlier values (nilai ekstrem)


## **2.1 Missing Values**

```python
# Persentase missing values per kolom
missing_pct = pd.DataFrame(
    df.isnull().sum() / len(df) * 100,
    columns=['Null Ratio (%)']
).round(2)
missing_pct['Jumlah Null'] = df.isnull().sum()
print(missing_pct)
```

**Interpretasi Missing Values:**
- Kolom `director` memiliki missing value tertinggi sebesar **~29.9%** — hal ini wajar karena banyak konten TV Show yang tidak memiliki satu sutradara spesifik.
- Kolom `cast` dan `country` masing-masing memiliki ~9.4% nilai kosong.
- Kolom penting seperti `title`, `type`, `listed_in`, `release_year`, dan `description` **tidak memiliki missing value (0%)** — data inti aman untuk dianalisis.


## **2.2 Inconsistent Values**

```python
# Cek nilai unik pada kolom rating — ada nilai aneh berupa durasi menit
print("Unique values di kolom 'rating':")
print(df['rating'].unique())
print("\nNilai tidak valid (seharusnya di kolom 'duration'):")
print(df[df['rating'].isin(['74 min', '84 min', '66 min'])][['title', 'type', 'rating', 'duration']])
```

**Temuan:** Terdapat 3 baris di kolom `rating` yang berisi nilai durasi seperti `'74 min'`, `'84 min'`, `'66 min'`. Nilai-nilai ini seharusnya berada di kolom `duration`. Ini adalah contoh **dirty data** yang perlu dibersihkan.


## **2.3 Duplicate Values**

```python
duplikat = df[df.duplicated()]
print(f"Jumlah baris duplikat: {len(duplikat)}")
print("✅ Tidak ada data duplikat!" if len(duplikat) == 0 else duplikat)
```

## **2.4 Outlier Values**

```python
# Deteksi outlier pada kolom release_year menggunakan metode IQR
q1 = df['release_year'].quantile(0.25)
q3 = df['release_year'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = df[(df['release_year'] < lower) | (df['release_year'] > upper)]
pct = len(outliers) / len(df) * 100

print(f"Batas bawah IQR : {lower:.0f}")
print(f"Batas atas IQR  : {upper:.0f}")
print(f"Jumlah outlier  : {len(outliers)} ({pct:.2f}%)")
print(f"\nRange release_year: {df['release_year'].min()} - {df['release_year'].max()}")
print("\nCatatan: Outlier berupa film-film lawas (1925-1970an) — valid secara konteks, tidak dihapus.")
```

# **3. Preprocessing (Data Cleaning)**

Berdasarkan hasil verifikasi kualitas data, langkah-langkah cleaning yang dilakukan:
1. Perbaiki nilai salah di kolom `rating` (pindahkan nilai durasi ke kolom yang benar)
2. Konversi kolom `date_added` ke format datetime
3. Ekstrak fitur `year_added` dan `month_added`
4. Isi missing values `director` dan `country` dengan `'Unknown'`
5. Ekstrak negara dan genre utama dari kolom multi-nilai


```python


df_clean = df.copy()

# 1. Fix nilai salah di kolom rating
mask_wrong = df_clean['rating'].isin(['74 min', '84 min', '66 min'])
df_clean.loc[mask_wrong, 'duration'] = df_clean.loc[mask_wrong, 'rating']
df_clean.loc[mask_wrong, 'rating'] = np.nan


# 2. Konversi date_added ke datetime
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'].str.strip(), format='%B %d, %Y', errors='coerce')
df_clean['year_added'] = df_clean['date_added'].dt.year
df_clean['month_added'] = df_clean['date_added'].dt.month


# 3. Isi missing values
df_clean['director'] = df_clean['director'].fillna('Unknown')
df_clean['country'] = df_clean['country'].fillna('Unknown')
df_clean['rating'] = df_clean['rating'].fillna('NR')


# 4. Ekstrak primary country dan primary genre
df_clean['primary_country'] = df_clean['country'].str.split(',').str[0].str.strip()
df_clean['primary_genre'] = df_clean['listed_in'].str.split(',').str[0].str.strip()


print(f"\nShape setelah cleaning: {df_clean.shape}")
print(f"Missing values tersisa:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
```
